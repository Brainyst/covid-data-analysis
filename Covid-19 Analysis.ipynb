## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"
```

## Importing the dataset using Pandas from above mentioned url

```python
df = pd.read_csv(url)
```

## Low Level Data Understanding:

```python
# Find count of unique values in location column.
unique_locations = df["location"].nunique()
print(unique_locations)
```

```python
# Find which continent has maximum frequency using values counts.
max_freq_cont = df["continent"].value_counts().idxmax()
print(max_freq_cont)
```

```python
# Find maximum & mean value in 'total_cases'.
max_total_cases = df["total_cases"].max()
mean_total_cases = df["total_cases"].mean()
print("Max Total Cases:",max_total_cases)
print("Mean Total Cases:",mean_total_cases)
```

```python
# Find 25%,50% & 75% quartile value in 'total_deaths'.
df["total_deaths"].quantile([0.25, 0.50, 0.75]).round(2)
```

```python
#  Find which continent has maximum 'human_development_index'.
max_hdi_continent = df.groupby("continent").agg({'human_development_index':'max'})
max_hdi_continent.sort_values(by='human_development_index', ascending=False).head(1)
```

```python
# Find which continent has minimum 'gdp_per_capita'.
min_gdp_per_capita = df.groupby("continent").agg({'gdp_per_capita':'min'})
min_gdp_per_capita.head(1)
```

## Filter the dataframe with only this columns 
['continent','location','date','total_cases','total_deaths','gdp_per_c

human_development_index'] and update the data frame.

```python
df =  df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
df
```

## Data Cleaning

```python
# Remove all duplicates observation
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Find missing values in all columns.
df.isnull().sum()
```

```python
# Remove all observations where continent column value is missing
df = df.dropna(subset=['continent'])
```

```python
df
```

```python
# Fill all missing values with 0
df = df.fillna(0)
df
```

```python
df.isnull().sum()
```

## Date time format:

```python
# Convert date column in datetime format using pandas.to_datetime
df["date"] = pd.to_datetime(df["date"])
```

```python
df.dtypes
```

```python
# Create new column month after extracting month data from date column
df["month"] = pd.DatetimeIndex(df["date"]).month
```

```python
df.head()
```

## Data Aggregation:

```python
# Find max value in all columns using groupby function on 'continent' column
df.groupby("continent").max().reset_index()
```

```python
# Store the result in a new dataframe named 'df_groupby'.
df_groupby = df.groupby("continent").max().reset_index()
```

```python
df_groupby
```

## Feature Engineering :

```python
#  Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby["total_deaths_to_total_cases"] = df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```

## Data Visualization :

```python
# Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
plt.figure(figsize=(6,4))
sns.distplot(df["gdp_per_capita"])
plt.title("Distribution of GDP per Capita")
plt.xlabel("GDP per Capita")
plt.ylabel("Density")
```

```python
# Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
plt.figure(figsize=(6,6))
sns.jointplot(data=df_groupby, x="total_cases",y="gdp_per_capita")
plt.xlabel("Total Cases")
plt.ylabel("GDP per Capita")
plt.show()
```

```python
# Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
```

```python
# Plot a bar plot of 'continent' column with 'total_cases' .
plt.figure(figsize=(6,6))
sns.catplot(data=df_groupby, x="continent", y="total_cases", kind='bar')
plt.xticks(rotation=45, ha='right')
plt.title("Total COVID-19 Cases by Continent")
plt.xlabel('Continent')
plt.ylabel('Total COVID-19 Cases')
plt.show()
```
